```python

import random
import numpy as np
import matplotlib.pyplot as plt
import statistics
import itertools
from Crypto.Cipher import DES
from Crypto.Cipher import DES3
```

```python


sysrandom = random.SystemRandom()
randArray = []

def HammingDistance(bits0,bits1):
    return bin(bits0 ^ bits1).count("1")



def randomBitFlip(bits0):
  bitnum = random.randint(0,63)
  bits1 = bits0 ^ (1 << bitnum)
  randArray.append(bits1)
        

randArray.append(sysrandom.getrandbits(64))


for i in range(0,10):
  randomBitFlip(randArray[i])
    

hamming_d1 = []
hamming_d2 = []

key1 = '01234567'
key2 = '01234568'
des1 = DES.new(key1, DES.MODE_ECB)
des2 =DES.new(key2,DES.MODE_ECB)
for i in range(0,10):
    ciphertext1 = des1.encrypt(randArray[i].to_bytes(8,'little'))
    ciphertext2 = des1.encrypt(randArray[i+1].to_bytes(8,'little'))
    d = HammingDistance(int.from_bytes(ciphertext1,'little'),int.from_bytes(ciphertext2,'little'))    
    hamming_d1.append(d)

for i in range(0,10):
    ciphertext1 = des2.encrypt(randArray[i].to_bytes(8,'little'))
    ciphertext2 = des2.encrypt(randArray[i+1].to_bytes(8,'little'))
    d = HammingDistance(int.from_bytes(ciphertext1,'little'),int.from_bytes(ciphertext2,'little'))    
    hamming_d2.append(d)

plt.plot(hamming_d1)
plt.title("Avalanche Effect on DES\n(Number of plaintext bits flipped=1)\n Key1")
plt.xlabel("DES rounds")
plt.ylabel("Hamming Distance")
plt.xlim(0,10)
plt.ylim(0,50)
plt.show()

plt.plot(hamming_d2)
plt.title("Avalanche Effect on DES\n(Number of plaintext bits flipped=1)\n Key2")
plt.xlabel("DES rounds")
plt.ylabel("Hamming Distance")
plt.xlim(0,10)
plt.ylim(0,50)
plt.show()
```
